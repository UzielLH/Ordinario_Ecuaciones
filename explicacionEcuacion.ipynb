# GRÁFICA SOBRE LA RESISTENCIA DEL AIRE #

La ecuación diferencial que describe la resistencia del aire es fundamental en la física y la ingeniería, especialmente en el estudio del movimiento de objetos en fluidos como el aire. Esta ecuación surge de consideraciones sobre las fuerzas que actúan sobre un objeto en movimiento a través del aire, y se expresa típicamente como una ecuación diferencial de segundo orden. La forma más común de esta ecuación se deriva de la ley de Newton para el movimiento, donde la fuerza de arrastre (resistencia del aire) es proporcional al cuadrado de la velocidad del objeto con respecto al aire. Esta relación no lineal conduce a una ecuación diferencial de segundo orden que puede resolverse para encontrar la velocidad o la posición del objeto en función del tiempo. La solución de esta ecuación es esencial en el diseño de vehículos, la aerodinámica, la predicción de trayectorias de proyectiles y en una variedad de otros campos donde se estudian objetos en movimiento a través de medios fluidos.



La ecuación diferencial que modela la resistencia del aire es:


$$m\frac{dv}{dt} = mg - kv$$


Donde:
- \( v \) es la velocidad,
- \( t \) es el tiempo,
- \( g \) es la gravedad,
- \( k \) es la constante de resistencia del aire, y
- \( m \) es la masa del objeto.


# Explicación del código #

## Librerias a ocupar ##

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

-numpy es una librería para cálculos numéricos en Python. Se utiliza aquí para crear arreglos y realizar operaciones numéricas.

-matplotlib.pyplot es una librería para crear visualizaciones en 2D en Python. Se usa aquí para graficar los resultados.

-scipy.integrate.odeint es una función de la librería SciPy que resuelve ecuaciones diferenciales ordinarias.

## Definición de la función ##

```python
# Función que representa la ecuación diferencial
def modelo(v, t, g, k, m):
    """
    Ecuación diferencial que modela el movimiento de un objeto bajo la influencia de la gravedad y la resistencia del aire.

    Parámetros:
    v: Velocidad
    t: Tiempo
    g: Gravedad (m/s^2)
    k: Constante de resistencia del aire (Ns/m)
    m: Masa (kg)
    """
    dvdt = g - (k/m) * v
    return dvdt
```

Esta función representa la ecuación diferencial que describe el movimiento de un objeto bajo la influencia de la gravedad y la resistencia del aire. En este caso se despeja para que quede solo el diferencial

## Solicitud de parámetros ##

```python
# Pedir al usuario los valores de los parámetros
try:
    v0 = float(input("Ingresa la velocidad inicial (m/s): "))
    g = float(input("Ingresa el valor de la gravedad (m/s^2): "))
    m = float(input("Ingresa el valor de la masa (kg): "))
    k = float(input("Ingresa el valor de la constante de resistencia del aire (Ns/m): "))

except ValueError:
    print("Error: Ingresa valores númericos")
    exit()
```

Estas líneas solicitan al usuario que ingrese los valores de la velocidad inicial, la gravedad, la masa y la constante de resistencia del aire, y los convierten en números de punto flotante. Si el usuario ingresa un valor no numérico, se muestra un mensaje de error y el programa termina.

## Resolución de la ecuación diferencial ##

```python
# Resolver la ecuación diferencial
t = np.linspace(0, 50, 200)  # Rango de tiempo
sol = odeint(modelo, v0, t, args=(g, k, m))
```

Estas líneas generan un rango de tiempo de 0 a 50 segundos con 200 puntos y resuelven la ecuación diferencial usando odeint. Los resultados se almacenan en sol.

## Creación de la gráfica ##

```python
# Graficar
fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('Tiempo (s)')
ax1.set_ylabel('Velocidad (m/s)', color=color)
ax1.plot(t, sol, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Resistencia del aire (N)', color=color)
resistencia = k * sol.flatten()  # Calcula la resistencia del aire
ax2.plot(t, resistencia, color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Solución de la ecuación diferencial mdv/dt = mg - kv')
plt.grid(True)
ax1.grid(True)  # Activa la cuadrícula en los ejes ax1
ax2.grid(True)  # Activa la cuadrícula en los ejes ax2
plt.show()
```

-Se crea una figura y un conjunto de ejes (ax1) usando plt.subplots().

-Se grafican los resultados de la velocidad en función del tiempo en ax1 y se establecen las etiquetas de los ejes.

-Se crea un segundo conjunto de ejes (ax2) compartiendo el eje x con ax1, para graficar la resistencia del aire en función del tiempo.

-Se grafica la resistencia del aire en ax2 y se establecen las etiquetas de los ejes.

-Se establece el título de la gráfica y se activa la cuadrícula en ambos conjuntos de ejes (ax1 y ax2).

-Finalmente, se muestra la gráfica.
